```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
```

```python
# Scatter plot for age vs. charges
plt.figure(figsize=(10, 6))
plt.scatter(insurance_data['age'], insurance_data['charges'], alpha=0.5)
plt.title('Age vs. Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Reshape data for the model
X = insurance_data['age'].values.reshape(-1, 1)  # Features
y = insurance_data['charges'].values  # Target variable

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict charges for a 70-year-old
age_70 = np.array([[70]])  # Age needs to be in the same format as X
prediction = model.predict(age_70)
print(f"Predicted charges for a 70-year-old: ${prediction[0]:.2f}")
```

```python
# Generate predictions for the linear model for all ages
age_range = np.arange(X.min(), X.max()).reshape(-1, 1)
charges_predictions = model.predict(age_range)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.5, label='Actual charges')  # Original data
plt.plot(age_range, charges_predictions, color='red', label='Best-fit line')  # Best-fit line
plt.title('Age vs. Charges with Linear Regression Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend()
plt.show()
```
